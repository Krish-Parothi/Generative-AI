#### Vector Store Retriever

```python
from langchain_classic.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.documents import Document
```

```python
document = [
    Document(page_content="LangChain helps developers build LLM Application Easily."),
    Document(page_content="Chroma is a vector database optimized for LLM-based search."),
    Document(page_content="Embeddings Convert Text into Highly Dimensional Vectors."),
    Document(page_content="OpenAI provides powerful embedding models.")
]
```

```python
embedding_model = HuggingFaceEmbeddings()
```

```python
vector_store = Chroma.from_documents(
    documents=document,
    embedding=embedding_model,
    collection_name="my_collection"
)
```

```python
retreiver = vector_store.as_retriever(search_kwargs={"k":2})
```

```python
query = "What is Chroma used for?"
result = retreiver.invoke(query)
```

```python
for i, doc in enumerate(result):
    print(f"\n--- Result {i+1} ---")
    print(doc.page_content)
```

```python
result = vector_store.similarity_search(query,k=2)
```

```python
for i, doc in enumerate(result):
    print(f"\n--- Result {i+1} ---")
    print(doc.page_content)
```
